```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data1=pd.DataFrame(data,columns=['gender','Churn'])
```

```python
data1
```

# 1. Compare churn count with respect to gender

```python
plt.figure(figsize=(5,6))
data2=sns.countplot(x='gender',hue='Churn',data=data1)
plt.grid()
```

# 2. Find out how many female senior citizens there in the dataset

```python
data3=data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
```

```python
data3
```

# 2. Find out how many female senior citizens there in the dataset

```python
plt.figure(figsize=(4,6))
sns.countplot(x=data3['gender'])
plt.title('Female senior citizens ',fontsize=15)
```

# 3. Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(10,8))
plt.scatter(data['tenure'],data['TotalCharges'],s=10,c='red',marker='^')
plt.title('Tenure v/s Total charges',fontsize=15)
plt.xlabel('Tenure',fontsize=12)
plt.ylabel('Total Charges',fontsize=12)
```

# 4. Find out which contract is preferred by the senior citizen.


```python
data4=data[(data['SeniorCitizen']==1)]
```

```python
data4
```

```python
plt.figure(figsize=(6,8))
sns.countplot(x=data4['Contract'])
plt.grid()
plt.title('Contract Preferred by Senior citizen',fontsize=15,fontweight='bold')
```
