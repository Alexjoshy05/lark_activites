```python
#2. Examine the data by using visualizations.
```

```python
plt.figure(figsize = (15, 7))
sns.barplot(x=data['country'],y=data['Population in thousands (2017)'],data=data,order=data.sort_values('Population in thousands (2017)',ascending=False).country.head(20))
plt.xlabel("Country",fontsize=20)
plt.ylabel("Population",fontsize=20)
plt.title("Top 20 Countries based on Population",fontsize=25)
plt.xticks(rotation=75)
plt.show()   
```

```python
#3. Find out how many countries with respect to the region are available in the dataset 
```

```python
data.groupby('Region')[['country']].count()
```

```python
#4. On average which region has the highest 
#a. GDP per capita (current US$) 
```

```python
data2=data.groupby('Region')[['GDP per capita (current US$)']].mean()
```

```python
data2
```

```python
data2.sort_values(by='GDP per capita (current US$)',ascending=False)
```

```python
data2.nlargest(1,'GDP per capita (current US$)')
```

```python
#b. International trade: Imports (million US$) 
```

```python
data3=data.groupby('Region')[['International trade: Imports (million US$)']].mean()
data3
```

```python
data3.nlargest(1,'International trade: Imports (million US$)')
```

```python
#5. Find out which region has more consistent with respect to GDP per capita (current US$)
```

```python
data4=data.groupby('Region')[['GDP per capita (current US$)']].std()
data4
```
